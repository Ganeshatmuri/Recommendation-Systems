```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("movies.csv",usecols=['movieId','title'])
```

```python
movies
```

```python
movies.info()
```

```python
rating=pd.read_csv('ratings.csv',usecols=['userId', 'movieId', 'rating'])
```

```python
rating
```

```python
rating.info()
```

```python
df=pd.merge(rating,movies,on='movieId')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
movie_count=pd.DataFrame(df.groupby('title')['rating'].count().reset_index())
```

```python
movie_count
```

```python
movie_count.rename(columns={"rating":"total rating count"},inplace=True)
```

```python
movie_count
```

```python
totalrating=pd.merge(df,movie_count,on='title')
```

```python
totalrating
```

```python
movie_count.describe()
```

```python
pop_movie=50
```

```python
pop_rating=totalrating[totalrating['total rating count']>=pop_movie]
```

```python
pop_rating
```

```python
pop_rating.shape
```

```python
movie_features_df=pop_rating.pivot_table(columns='userId',index='title',values='rating').fillna(0)
movie_features_df
```

```python
from scipy.sparse import csr_matrix
```

```python
movie_features_matrix=csr_matrix(movie_features_df.values)
```

```python
movie_features_df.values
```

```python
movie_features_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
model_knn.fit(movie_features_matrix)
```

```python
movie_features_df.shape
```

```python
new=np.random.choice(movie_features_df.shape[0])
```

```python
new
```

```python
movie_features_df.iloc[new,:]
```

```python
movie_features_df.iloc[new,:].values.reshape(1,-1)
```

```python
distance,indices=model_knn.kneighbors(movie_features_df.iloc[new,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
indices
```

```python
movie_features_df.index[new]
```

```python
for i in range(0,len(distance.flatten())):
    if i==0:
        print("Recommendation for {}:\n".format(movie_features_df.index[new]))
    else:
        print(movie_features_df.index[indices.flatten()[i]],"with distance of",distance.flatten()[i])
```
