```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
books=pd.read_csv("BX-Books.csv",sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
users=pd.read_csv("BX-Users.csv",sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings=pd.read_csv("BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings['Book-Rating'].hist()
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

```python
users['Age'].hist(bins=np.arange(0,110,10))
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
count=ratings['User-ID'].value_counts()
```

```python
exe_ratings=ratings[ratings['User-ID'].isin(count[count>=200].index)]
```

```python
book=ratings['Book-Rating'].value_counts()
```

```python
exe_ratings=exe_ratings[exe_ratings['Book-Rating'].isin(book[book>=100].index)]
```

```python
exe_ratings
```

```python
combine=exe_ratings.merge(books,on='ISBN')
```

```python
combine.columns
```

```python
combine.drop(['Book-Author',
       'Year-Of-Publication', 'Publisher', 'Image-URL-S', 'Image-URL-M',
       'Image-URL-L'],1,inplace=True)
```

```python
combine
```

```python
book_count=pd.DataFrame(combine.groupby(['Book-Title'])['Book-Rating'].count().reset_index())
```

```python
book_count.rename(columns={"Book-Rating":"Total Rating"},inplace=True)
```

```python
book_count
```

```python
combine=combine.merge(book_count,on='Book-Title')
```

```python
combine
```

```python
popular_books=combine[combine['Total Rating']>=50]
```

```python
popular_books
```

```python
us_canada=popular_books.merge(users,on='User-ID')
```

```python
us_canada
```

```python
us_canada.drop("Age",1,inplace=True)
```

```python
us_canada=us_canada[us_canada['Location'].str.contains('usa|canada')]
```

```python
us_canada
```

# KNN

```python
from scipy.sparse import csr_matrix
```

```python
us_canada=us_canada.drop_duplicates(['User-ID','Book-Title'])
```

```python
us_canada
```

```python
us_canada_pivot=us_canada.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating').fillna(0)
```

```python
us_canada_pivot
```

```python
us_canada_matrix=csr_matrix(us_canada_pivot.values)
```

```python
us_canada_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
model_knn.fit(us_canada_matrix)
```
