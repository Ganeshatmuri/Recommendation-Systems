```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head(3)
```

```python
credits.isnull().sum()
```

```python
movies.isnull().sum()
```

```python
credits.shape
```

```python
movies.shape
```

```python
credits.rename(columns={"movie_id":"id"},inplace=True)
```

```python
movies=pd.merge(movies,credits,on='id')
```

```python
movies.shape
```

```python
movies.columns
```

```python
movies.drop(['homepage','title_x','title_y','status','production_countries'],1,inplace=True)
```

```python
movies.shape
```

```python
movies.head(3)
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
v=movies['vote_count']
v
```

```python
R=movies['vote_average']
R
```

```python
C=movies['vote_average'].mean()
C
```

```python
m=movies['vote_count'].quantile(0.7)
m
```

```python
movies['Weighted_Average']=((R*v)+(C*m))/(v+m)
```

```python
movies.head(3)
```

```python
movies_sorted=movies.sort_values(by='Weighted_Average',ascending=False,ignore_index=True)
```

```python
movies_sorted[['original_title','vote_count','vote_average','Weighted_Average','popularity']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies['Weighted_Average'],y=movies['original_title'].head(10))
```

```python
popularity=movies.sort_values(by="popularity",ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y=popularity["original_title"].head(10),x=popularity["popularity"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()
```

```python
scaled=scaling.fit_transform(movies[['Weighted_Average','popularity']])
```

```python
movie_normalized=pd.DataFrame(scaled,columns=['norm_Weighted_Average','norm_popularity'])
```

```python
movie_normalized['score']=movie_normalized['norm_Weighted_Average']*0.5+movie_normalized['norm_popularity']*0.5
```

```python
movie_normalized
```

```python
movies=pd.concat([movies,movie_normalized],1)
```

```python
movies
```

```python
movies.sort_values('score',ascending=False,inplace=True,ignore_index=True)
```

```python
movies[['original_title','norm_Weighted_Average','norm_popularity','score']].head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y=movies['original_title'].head(10),x=movies['score'])
```
